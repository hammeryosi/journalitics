```python
def getTopNameCount(startDate, endDate):
    keep = 20
    data = pd.read_sql_query("select * from name_counts where date >= '" + 
                         startDate + "' and date <= '" + endDate + "'", engine)
    jours = pd.unique(data['journal'])
    groups = data.groupby(['journal', 'name']).groups
    results = {}
    for j in jours:
        res = {}
        jGroups = [g for g in groups.keys() if g[0] == j]
        nameCount = [(g[1], sum(data['count'][groups[g]])) for g in jGroups]
        nameCount = sorted(nameCount, key=lambda x: x[1], reverse=True)[:20]
        res['names'] = [x[0] for x in nameCount]
        res['count'] = [int(x[1]) for x in nameCount]
        results[j] = res
    return results
```
